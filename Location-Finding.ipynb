```python
def time_to_minutes(time_str):
    try:
        time_obj = datetime.strptime(time_str, "%I:%M %p")
        return time_obj.hour * 60 + time_obj.minute
    except ValueError:
        return None

data['sunrise_minutes'] = data['sunrise'].apply(time_to_minutes)
data['sunset_minutes'] = data['sunset'].apply(time_to_minutes)
data['moonrise_minutes'] = data['moonrise'].apply(time_to_minutes)
data['moonset_minutes'] = data['moonset'].apply(time_to_minutes)

d['sunrise_minutes'] = d['sunrise'].apply(time_to_minutes)
d['sunset_minutes'] = d['sunset'].apply(time_to_minutes)
d['moonrise_minutes'] = d['moonrise'].apply(time_to_minutes)
d['moonset_minutes'] = d['moonset'].apply(time_to_minutes)

```

```python
# Convert Time Zone to UTC Offset
def get_utc_offset(tz_name):
    tz = timezone(tz_name)
    offset = tz.utcoffset(datetime.now()).total_seconds() / 3600
    return offset

data['utc_offset'] = data['timezone'].apply(get_utc_offset)

d['utc_offset'] = d['timezone'].apply(get_utc_offset)

```

# Model

```python
# Haversine function to calculate the distance between two points
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    R = 6371.0  # km
    distance = R * c
    return distance
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = train_df.drop(columns=['latitude', 'longitude'])  # all columns except latitude and longitude
y = train_df[['latitude', 'longitude']]  # latitude and longitude
scaler = StandardScaler()
X = scaler.fit_transform(X)
X = pd.DataFrame(X)


X_val = val_df.drop(columns=['latitude', 'longitude'])  
y_val = val_df[['latitude', 'longitude']]  
X_val = scaler.transform(X_val)
```

# Randomtree

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

rf_model = RandomForestRegressor(n_estimators=200, max_depth=20, random_state=42)

rf_model.fit(X, y)

# Make predictions
y_pred_rf = rf_model.predict(X)

haversine_distances_rf = []
for i in range(len(y_pred_rf)):
    distance = haversine(y['latitude'].iloc[i], y['longitude'].iloc[i], y_pred_rf[i, 0], y_pred_rf[i, 1])
    haversine_distances_rf.append(distance)

avg_distance_rf = np.mean(haversine_distances_rf)
print(f"Average Haversine Distance with Random Forest: {avg_distance_rf} km")
```
